#Introduction

- ML is a subset of AI.
- Name derived from the concept it deals with "Construction and study of systems that can learn from data."

In Machine Learning (ML), there are 3 main types, and each one solves a different kind of problem.

##1. Supervised Learning
**Definition:** The model learns from labeled data â€” where input and correct output are already known.

**Example:**

You have data of students' study hours (input) and their marks (output).
The model learns the relationship and predicts marks for new students.

**Goal:** Find patterns or groupings in the data.

**Types of Supervised Learning:**

| Subtype            | Task               | Example                |
| ------------------ | ------------------ | ---------------------- |
| **Regression**     | Predict a number   |  House Price Prediction, Temp Forecast|
| **Classification** | Predict a category | Churn Prediction, Disease deetction |



---

###Supervised Learning Steps:

1. **Collect Data**  
   Example: Marks of students based on study hours

2. **Split the Data**
   - **Training Set** â†’ Used to train the model (usually 70â€“80%)
   - **Test Set** â†’ Used to test how well the model performs on new/unseen data (20â€“30%)

3. **Train the Model**  
   - Fit the model using the **training data**
   - It learns the pattern (e.g., how marks change with study hours)

4. **Test the Model**  
   - Give the model the **test data** (without answers)
   - Compare its predictions with actual answers

5. **Evaluate Performance**
   - Use metrics like accuracy, MAE, RMSE, F1-score (depending on the task)


##2. Unsupervised Learning
**Definition:** The model learns from unlabeled data â€” no correct answers given.

**Example:**
You have customer data (age, income), and the model groups similar customers together (clusters), without knowing who belongs where.

**Goal:** Find patterns or groupings in the data.

**Types of Unsupervised Learning:**

| Subtype                       | Task                      | Example                                  |
| ----------------------------- | ------------------------- | ---------------------------------------- |
| **Clustering**                | Group similar data points | Customer segmentation                    |
| **Association Rule Learning** | Find item patterns        | Market basket analysis (Amazon/Flipkart) |

##3. Reinforcement Learning
**Definition:** The model learns by trial and error â€” it gets rewards or penalties for actions taken.

**Example:**
A robot learns to walk â€” every time it moves correctly, it gets a reward; wrong moves give a penalty. Over time, it learns to walk well.

**Goal:** Learn to make a sequence of decisions.

##Feature Scaling
Feature Scaling is the process of adjusting the range of values (magnitude) of different features (columns) in a dataset.

Why Is It Needed?

In ML, many algorithms (like KNN, SVM, Linear Regression) use Euclidean distance to measure similarity.

If features have different units or magnitudes, one feature can dominate others.

ðŸ”§ Example:

Sepal Length = 4.5

Sepal Width = 0.3

â†’ The length value is much higher â€” so it will dominate the calculation, even if width is more important.

So, we scale the features to bring them to similar ranges, which improves model accuracy and fairness.

---
| Method                    | Description                                                       |
| ------------------------- | ----------------------------------------------------------------- |
| 1. **Standardisation**    | Converts data to have **mean = 0** and **std = 1**                |
| 2. **Mean Normalization** | Centers data around **mean**                                      |
| 3. **Min-Max Scaling**    | Scales data between **0 and 1** (or other fixed range)            |
| 4. **Vector Scaling**     | Normalizes values to make vector length = 1 (used in ML like NLP) |


### **Statistically, standardization & normalization**

â†’ These are usually used for **feature scaling**.

---

### 1ï¸âƒ£ **Standardization**

$$
X' = \frac{X - \mu}{\sigma}
$$

Where:

* $\mu$ = Mean
* $\sigma$ = Standard Deviation

---

### 2ï¸âƒ£ **Normalization**

$$
X' = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
$$

---

### Example Table:

| Height | Weight | Ht (Height Ã· 1000) | Ht' (Normalized Height) |
| ------ | ------ | ------------------ | ----------------------- |
| 172    | 80     | 0.172              | 0.6                     |
| 170    | 80     | 0.170              | 0.5                     |
| 180    | 85     | 0.180              | 1                       |
| 160    | 70     | 0.160              | 0                       |
| 165    | 72     | 0.165              | 0.25                    |

---

### Additional Notes:

* $X_{\max} = 180$
* $X_{\min} = 160$

---

### Calculations:

1. For 172:

   $$
   \frac{172 - 160}{180 - 160} = \frac{12}{20} = 0.6
   $$

2. For 170:

   $$
   \frac{170 - 160}{180 - 160} = \frac{10}{20} = 0.5
   $$

3. For 165:

   $$
   \frac{165 - 160}{180 - 160} = \frac{5}{20} = 0.25
   $$

### **K-NN (K-Nearest Neighbours)**

```
       |
    B(8,6)     *
       |    /
       |   /
       |  /
       | /    â† Distance between A and B
       |/
A(2,3)*--------------------
```


#### Distance Calculation

To calculate distance between points A and B:

* $A = (2, 3)$
* $B = (8, 6)$

$$
\text{Distance (AB)} = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2} = \sqrt{6^2 + 3^2} = \sqrt{45} â‰ˆ 6.7
$$

This is called **Euclidean Distance**.

---

###  **K-NN Example (Binary Classification)**

```
     x   x     x
         x   x    
      x     x   
         O     
      1   1   1     
         1    
```


####  Scenario:

* Symbols:

  * `x`: Cancer Positive
  * `1`: Cancer Negative
  * `o`: New patient

#### Question:

Is the new patient diagnosed with cancer?

#### Steps:

1. Choose an **odd value for K** (e.g., 3, 5, 7) to avoid tie.
2. If **K = 3**, check the **3 nearest neighbours**.
3. If among the 3 nearest:

   * 2 are `x` and 1 is `1` â†’ new patient is `x` (Cancer Positive)

##K-Means Clustering

````python
##Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

df_sc = sc.fit_transform(df)
````

````python
## Elbow Method

from sklearn.cluster import KMeans

# Create an empty list to store WCSS (within-cluster sum of squares)
wcss = []

# Try different numbers of clusters from 1 to 10
for i in range(1,11):
  # Initialize KMeans with i clusters
  # 'k-means++' helps to choose better initial centroids
  kmeans = KMeans(n_clusters = i, init = "k-means++")

  # Fit KMeans to your data (should be scaled: df_sc)
  kmeans.fit(df_sc)

  # Save the inertia (WCSS) value for this number of clusters
  wcss.append(kmeans.inertia_)

# Plot the WCSS values to visualize the "elbow"
plt.plot(range(1,11), wcss, marker = "o")
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
````

````python
#STATIONARITY CHECK
#Ho: TS is non-Stationary
#H1: TS is stationary
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):

    #Determing rolling statistics
    #rolmean = pd.rolling_mean(timeseries, window=12)
    rolmean = pd.Series(timeseries).rolling(window=12).mean()
    #rolstd = pd.rolling_std(timeseries, window=12)
    rolstd = pd.Series(timeseries).rolling(window=12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)

    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key,value in dftest[4].items():
      dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
````

````python
test_stationarity(ts)
````

If p-value > 0.05 --> H0 is True which means it is not a stationary

If p-value <= 0.05 --> H1 is True which means it is a stationary

Here, p-value is 0.99. Therefore the above data is not stationary

## Random Search CV

````python
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
max_depth = [int(x) for x in np.linspace(start = 10, stop = 110, num = 11)]
min_samples_leaf = [1,2,4,10,20,50,100]
min_samples_split = [2,3,4,5,8,10,20,50,100,200]
bootstrap = [True, False]

# Create the Random Grid
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
````

````python
print("n_estimators: ", n_estimators)
print("max_depth: ", max_depth)
````

On each iteration, the algorithm will choose a different combination of the features. Altogether there are 15,400 combinations. However the benifit of the random search is that we are not trying every combination , but selecting at random to sample a wide range of values.

````python
## Use the random grid to search for the best hyper parameters
## first create the base model to tune

rf = RandomForestClassifier()
# random search of the parameters using 3 fold cross validations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, n_jobs = -1)

#Now fit the random search model
rf_random.fit(X_train, y_train)
````

````python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    accuracy = accuracy_score(test_labels, predictions)
    print('Model Performance')
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    return accuracy
````

````python
base_model = RandomForestClassifier(n_estimators = 5, random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)
````

````python
best_random = rf_random.best_estimator_
print(best_random)
````

````python
random_accuray = evaluate(best_random, X_test, y_test)
````

````python
print("Improvement of {:0.2f}%.".format((random_accuray-base_accuracy)/base_accuracy))
````
